# [Hubble Space Telescope (hst)](https://registry.opendata.aws/hst/)

using [astropy.io.fits](https://docs.astropy.org/en/stable/io/fits/)

```python
import os
import numpy as np
from abcli import fullname
from abcli import string
from abcli.plugins import seed
from hubble import NAME, VERSION
from hubble.ingest import ingest
from abcli.logger import logger

logger.info(f"{NAME}.{VERSION}.{fullname()}, built on {string.pretty_date()}")
```

```python
dataset_name = "hst"
hubble_object_name = "public/u4ge/u4ge0106r"
object_name = "2023-09-15-19-34-36-01196"

do_download = False
do_upload = False
```

```python
HOME=os.getenv("HOME")
%env abcli_object_name={object_name}
%env abcli_object_root={HOME}/storage/abcli
%env abcli_object_path={HOME}/storage/abcli/{object_name}
```

```python
%%bash -s $do_download $object_name
source ~/git/awesome-bash-cli/bash/abcli.sh in_notebook

do_download=$1
object_name=$2

if [[ "$do_download" == True ]] ; then
    abcli download $object_name
fi
```

---

```python
ingest(
    object_name,
    dataset_name,
    hubble_object_name,
    plot=True,
)
```

---

```python
%%bash -s $do_upload $object_name
source ~/git/awesome-bash-cli/bash/abcli.sh in_notebook

do_upload=$1
object_name=$2

abcli select $object_name
abcli tag set . built_by_notebook

if [[ "$do_upload" == True ]] ; then
    abcli upload
fi
```
